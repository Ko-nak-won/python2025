# DataFrame에 있는 데이터 그룹핑해서 보기

```python
import pandas as pd
```

#### groupby
  + 데이터를 그룹화(groupping)
  + SQL의 group by와 동일한 개념

```python
marathon_2017 = pd.read_csv("../marathon_data/marathon_results_2017.csv")
marathon_2017.head()
```

#### 그룹핑된 데이터의 groups 속성
 - 각 그룹과 그룹에 속한 index를 dictionary로 표현

#### 단일 column으로 그룹핑

```python
gender_group = marathon_2017.groupby('M/F')
gender_group
```

```python
gender_group.groups  # dictionary로 표현
```

#### 그룹핑된 데이터와 함께 사용되는 통계함수
 - NaN 값은 제외하고 연산
 - count - 데이터 개수 
 - sum   - 데이터의 합
 - mean, std, var - 평균, 표준편차, 분산
 - min, max - 최소, 최대값

```python
gender_group.count()
```

```python
gender_group.mean()['Age']
```

```python
marathon_2017[marathon_2017['Country']=='KEN'].groupby('M/F').mean()['Age']
```

#### 복수개의 column으로 그룹핑
 - groupby에 column 리스트를 전달
 - 통계함수를 적용한 결과는 multiindex를 갖는 DataFrame

```python
marathon_2017.groupby(['Country', 'M/F']).mean()
```

```python
marathon_2017.groupby(['Country', 'M/F']).mean()['Age']
```

```python
marathon_2017.groupby(['Country', 'M/F']).mean().loc[('VEN', 'F')]
```

```python
marathon_2017.groupby(['Country', 'M/F']).mean().index
```

#### index를 이용한 그룹핑
  - index가 있는 경우, groupby 함수에 level 사용 가능
  - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
  - **set_index** 함수
    - column을 index 레벨로 변경
  - **reset_index** 함수
    - index 초기화

```python
marathon_2017.head()
```

```python
marathon_2017.set_index('Country')  # Country 를 인덱스로 지정
```

```python
marathon_2017.set_index(['Country', 'M/F'])  # Country 와 M/F 을 멀티 인덱스로 지정
```

```python
marathon_2017.set_index(['Country', 'M/F']).reset_index()  # 기본 인덱스로 초기화
```

```python
# index가 한개있으면 level이 0부터 시작. M/F 로 groupby 하여 Age의 평균을 구함.
marathon_2017.set_index('M/F').groupby(level=0).mean()['Age']
```

#### 복수개의 index으로 그룹핑

```python
marathon_2017.set_index(['Country', 'M/F']).groupby(level=[0, 1]).mean()['Age']
```
